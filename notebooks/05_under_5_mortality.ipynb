```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

# df['under_5_mortality_rate'].unique ()

df=pd.read_csv(r'C:\Users\USER\Desktop\mch_county_dashboard\data\processed\mch_county_cleaned.csv')
print('Dataset loaded:',df.shape)

df['under_5_mortality_rate'] =pd.to_numeric(df['under_5_mortality_rate'],errors= 'coerce')

# This converts numbers to float and non numeric values to NaNnon-numeric values to NaN

df = df.dropna(subset=['under_5_mortality_rate'])
print('values converted successfully')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,14))

u5_data = df.set_index('county')[['under_5_mortality_rate']]
sns.heatmap(
    u5_data,
    cmap='Oranges',
    linewidths=0.5,
    annot=True,
    fmt='.0f'
)
plt.title('County-Level Under-Five Mortality Rate(per 1,000)',fontsize=14)
plt.ylabel('County')
plt.xlabel('Indicator')
plt.tight_layout()
plt.show()
```
