```python
# Import and load processed data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')

df = pd.read_csv(r'C:\Users\USER\Desktop\mch_county_dashboard\data\processed\mch_county_cleaned.csv')

print('Dataset loaded:',df.shape)
df.head()
      
```

```python
# County heatmap: service coverage
import matplotlib.pyplot as plt

plt.figure(figsize=(10,14))

heatmap_data = df.set_index('county')[
    [
        'skilled_birth_attendance_pct',
        'antenatal_4_visits_pct',
        'facility_delivery_pct',
        'postnatal_care_48_hours'
    ]
]

sns.heatmap(
    heatmap_data,
    cmap='RdYlGn',
    linewidths=0.5,
    annot=True,
    fmt='.0f'
)

plt.title('County-Level Maternal Health Service Coverage (%)', fontsize=14)
plt.ylabel('County')
plt.xlabel('Indicators')
plt.tight_layout()
plt.show()
```
