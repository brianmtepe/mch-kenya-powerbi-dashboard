```python
import os
import pandas as pd 
import numpy as np
print('Working directory:',os.getcwd())

np.random.seed(42)

# Ensure data folder exists
os.makedirs(r'C:\Users\USER\Desktop\mch_county_dashboard\data', exist_ok=True)

print('Environment ready')
```

```python
counties = [
    'Mombasa','Kwale', 'Kilifi','Lamu','Tana River','Taita Taveta','Garissa',
    'Wajir','Mandera','Marsabit','Isiolo','Meru','Tharaka Nithi','Embu','Kitui',
    'Machakos','Makueni','Nyandarua','Nyeri','Kirinyaga','Muranga','Kiambu',
    'Turkana','West Pokot','Samburu','Trans Nzoia','Uasin Gishu','Elgeyo Marakwet',
    'Nandi','Baringo','Laikipia','Nakuru','Narok','Kajiado','Kericho','Bomet',
    'Kakamega','Vihiga','Bungoma','Busia','Siaya','Kisumu','Homabay','Migori',
    'Kisii','Nyamira','Nairobi'    
]
len(counties)    
```

```python
np.random.seed(42)  # reproducibility

data = {
    'county':counties,
    'skilled_birth_attendance_pct':np.random.randint(40,95, size = 47),
    'antenatal_4_visits_pct':np.random.randint(35,90, size = 47),
    'facility_delivery_pct':np.random.randint(45,95, size = 47),
    'postnatal_care_48_hours':np.random.randint(30,85, size = 47),
    'under_5_mortality_rate':np.random.randint(20,90, size = 47),  #per 1,000
    'maternal_mortality_ratio':np.random.randint(200,800, size = 47) #per 100,000
}
df_mch = pd.DataFrame(data)
df_mch.head()
```
